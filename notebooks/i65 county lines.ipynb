### Import neccesary packages

```python
from shapely.geometry import Point, LineString
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import folium
from folium.plugins import MarkerCluster
from folium.plugins import FastMarkerCluster
```

### Create the map

```python
x = -86.78069                                                          # create the map on which all the points will be plotted
y = 36.14848
area_center = [y, x]
i65_map = folium.Map(location =  area_center, zoom_start = 10.0)
```

### Create Robertson County i65 LineStrings

```python
robertson = pd.read_excel('../data/robertson2020.xlsx', skiprows = 11)                   # read in the excel file.
robertson.rename(columns  = {'Latest' : 'AADT'}, inplace = True)
robertson = robertson[['Loc ID', 'County', 'On', 'At', 'Dir','Directions', 'Latitude', 'Longitude', 'AADT', 'Latest Date']]
i65robertson = robertson.loc[(robertson['On'] == 'I-65') | (robertson['On'] == 'I0065')]    # locate both i24 and i24 ramp (I0024).ID
```

```python
i65robertson = i65robertson.loc[[103,42,1,23,25,98,15,19,102,10,12,99,4,6]]
```

```python
i65robertson['geometry'] =i65robertson.apply(lambda x: Point((float(x.Longitude),        # create geometry column.
                                                         float(x.Latitude))),          # points in this col necessary for Line.
                                        axis=1)
i65robertson['line'] = LineString(i65robertson.geometry.tolist())                        # create line column which concats all
i65robertson_geo = gpd.GeoDataFrame(i65robertson,                                        # points together for linestring.
                           crs = "EPSG:4326",
                           geometry = i65robertson['geometry'])
i65robertson_geo                                                                            # hopefully it gives no trouble
```

### Create Davidson County i65 LineStrings

```python
davidson = pd.read_excel('../data/davidson2020.xlsx', skiprows = 11)                   # read in the excel file.
davidson.rename(columns  = {'Latest' : 'AADT'}, inplace = True)
davidson = davidson[['Loc ID', 'County', 'On', 'At', 'Dir','Directions', 'Latitude', 'Longitude', 'AADT', 'Latest Date']]
i65davidson = davidson.loc[(davidson['On'] == 'I-65') | (davidson['On'] == 'I0065')]    # locate both i24 and i24 ramp (I0024).
i65davidson = i65davidson.sort_values(by = 'Latitude', ascending = False)                                    # sort values by lat for LineString.

i65davidson['geometry'] =i65davidson.apply(lambda x: Point((float(x.Longitude),        # create geometry column.
                                                         float(x.Latitude))),          # points in this col necessary for Line.
                                        axis=1)
i65davidson['line'] = LineString(i65davidson.geometry.tolist())                        # create line column which concats all
i65davidson_geo = gpd.GeoDataFrame(i65davidson,                                        # points together for linestring.
                           crs = "EPSG:4326",
                           geometry = i65davidson['geometry'])
i65davidson_geo                                                                            # hopefully it gives no trouble
```

### Create Williamson County i65 LineStrings

```python
williamson = pd.read_excel('../data/williamson2020.xlsx', skiprows = 11)                   # read in the excel file.
williamson.rename(columns  = {'Latest' : 'AADT'}, inplace = True)
williamson = williamson[['Loc ID', 'County', 'On', 'At', 'Dir','Directions', 'Latitude', 'Longitude', 'AADT', 'Latest Date']]
i65williamson = williamson.loc[(williamson['On'] == 'I-65') | (williamson['On'] == 'I0065')]    # locate both i24 and i24 ramp (I0024).
i65williamson = i65williamson.sort_values(by = 'Latitude', ascending = False)                                    # sort values by lat for LineString.

i65williamson['geometry'] =i65williamson.apply(lambda x: Point((float(x.Longitude),        # create geometry column.
                                                         float(x.Latitude))),          # points in this col necessary for Line.
                                        axis=1)
i65williamson['line'] = LineString(i65williamson.geometry.tolist())                        # create line column which concats all
i65williamson_geo = gpd.GeoDataFrame(i65williamson,                                        # points together for linestring.
                           crs = "EPSG:4326",
                           geometry = i65williamson['geometry'])
i65williamson_geo                                                                            # hopefully it gives no trouble
```

```python
i65 = pd.concat([i65robertson_geo, i65sumner_geo, i65davidson_geo, i65williamson_geo])
```

```python
i65.reset_index(inplace = True)
```

```python
i65['line string'] = LineString(i65.geometry.tolist())
i65
```
