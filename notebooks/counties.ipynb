```python
from shapely.geometry import Point
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import folium
from folium.plugins import MarkerCluster
from folium.plugins import FastMarkerCluster
```

```python
bedford = pd.read_excel('../data/bedford2020.xlsx',skiprows = 11 )
cheatham = pd.read_excel('../data/cheatham2020.xlsx',skiprows = 11)
davidson = pd.read_excel('../data/davidson2020.xlsx',skiprows = 11)
dickson = pd.read_excel('../data/dickson2020.xlsx',skiprows = 11)
montgomery = pd.read_excel('../data/montgomery2020.xlsx',skiprows = 11)
robertson = pd.read_excel('../data/robertson2020.xlsx',skiprows = 11)
rutherford = pd.read_excel('../data/rutherford2020.xlsx',skiprows = 11)
sumner = pd.read_excel('../data/sumner2020.xlsx',skiprows = 11)
williamson = pd.read_excel('../data/williamson2020.xlsx',skiprows = 11)
wilson = pd.read_excel('../data/wilson2020.xlsx',skiprows = 11)
```

```python
counties = pd.concat([bedford,cheatham,davidson,dickson,montgomery,robertson,rutherford,sumner,williamson,wilson])
```

```python
counties.rename(columns = {'Latest' : 'AADT', 'Latest Date' : 'Year'}, inplace = True)
counties['Year'] = pd.to_datetime(counties['Year'])
counties['Year'] = counties['Year'].dt.year
counties['Year'] = 2021
```

```python
counties.to_csv('counties.csv')
```
